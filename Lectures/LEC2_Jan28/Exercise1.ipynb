### Logarithmic Compression

```python
#### C = 0.1
import numpy as np
im1 = 0.1 * np.log10(1+image*255)
IP.plt.imshow(im1,cmap="gray")
```

```python
#### C = 10 
im2 = 10 * np.log10(1+image*255)
IP.plt.imshow(im2,cmap="gray")
```

```python
#### C = 1000
im3 = 1000 * np.log10(1+image*255)
IP.plt.imshow(im3,cmap="gray")
```

#### Bottom line of the story
* Changing 'C' doesn't do much but scale the intensity values of the image

### Gamma Transformation

```python
#alpha = 0.1
im4 = image**0.1
IP.plt.imshow(im4,cmap="gray")
```

```python
#alpha = 0.5
im5 = image**0.5
IP.plt.imshow(im5,cmap="gray")
```

```python
#alpha = 10
im6 = image**10
IP.plt.imshow(im6,cmap="gray")
```
