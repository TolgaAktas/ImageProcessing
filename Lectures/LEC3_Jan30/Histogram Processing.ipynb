## Isaac Image

```python
isaac = IP.imread("Isaac.jpg");
IP.imshow(isaac,aspect="auto")
```

```python
isaacGray = IP.rgb2gray(isaac)
isaacGray = IP.imresize(isaacGray,(256,256))
IP.imshow(isaacGray,aspect="auto",cmap="gray"); IP.plt.axis("off")
```

```python
isaacGray.max()
```

```python
logIsaac = np.log10(1+np.double(isaacGray))
IP.imshow(logIsaac,aspect='auto',cmap='gray')
```

```python
map = 100 * np.log10(IP.linspace(0,255,256)+1)
map.min()
```

```python
IP.plot(map)
```

```python
mymap = IP.zeros((256,3))
```

```python
#Scale colormap between 0 and 1
map = map - map.min()
map = map/map.max()
mymap[:,0] = map
mymap[:,1] = map
mymap[:,2] = map

map1 = IP.colormap(mymap)
```

```python
IP.imshow(isaacGray,aspect="auto",cmap=map1);IP.plt.axis("off")
```

```python
IP.subplot(1,3,1);IP.imshow(isaacGray,aspect='auto',cmap="gray");IP.plt.axis("off")
IP.subplot(1,3,2);IP.imshow(isaacGray,aspect='auto',cmap=map1);IP.plt.axis("off")
IP.subplot(1,3,3);IP.imshow(np.log10(1+np.double(isaacGray)),aspect="auto",cmap="gray");IP.plt.axis("off")
```

## River Image

```python
river = IP.imread("River.jpg")
river2 = IP.rgb2gray(river)
np.shape(river2)
```

```python
river3 = IP.imresize(river2,(256,256),"bilinear")
IP.imshow(river3,aspect="auto",cmap="gray");IP.plt.axis("off")
```

```python
n,bins,patches = plt.hist(river3.ravel(),bins=250,histtype='step',fill=True,color="black")
```

```python
n_norm = n/(256*256)
plt.plot(n_norm)
```

```python
cdf = IP.cumsum(n_norm)
np.shape(cdf)
plt.plot(cdf);plt.xlabel("In")
```

```python
mymap=IP.zeros((250,3))
mymap[:,0] = cdf
mymap[:,1] = cdf
mymap[:,2] = cdf

mymap = IP.colormap(mymap)
```

```python
IP.imshow(river3,aspect="auto",cmap=mymap);plt.axis('off')
```
