```python
#Import IP.py module for accessing other modules
ip_path = '~/Desktop/SPRING 2019/IP-ECE447/IP.py'

import os
import sys
sys.path.append(os.path.dirname(os.path.expanduser(ip_path)))

import IP
```

```python
isaacOriginal = IP.imread("Isaac.jpg");
```

```python
import skimage
from skimage.transform import resize
skimage.__version__
```

```python
isaac = IP.rgb2gray(isaacOriginal)
isaac = resize(isaac,(256,256),anti_aliasing=True)
IP.imshow(isaac,cmap='gray',aspect='auto')
```

```python
k1 = (1/(5*5))*IP.ones((5,5))
k2 = (1/(11*11) * IP.ones((11,11)))
k3 = (1/(15*15) * IP.ones((15,15)))
k4 = (1/(21*21) * IP.ones((21,21)))
k5 = (1/(51*51) * IP.ones((51,51)))
```

```python
I1 = IP.filter2D(isaac,k1,"same")
I2 = IP.filter2D(isaac,k2,"same")
I3 = IP.filter2D(isaac,k3,"same")
I4 = IP.filter2D(isaac,k4,"same")
I5 = IP.filter2D(isaac,k5,"same")
```

```python
IP.subplot(2,3,1);IP.imshow(isaac,aspect='auto',cmap="gray");IP.plt.axis("off")
IP.subplot(2,3,2);IP.imshow(I1,aspect='auto',cmap="gray");IP.plt.axis("off")
IP.subplot(2,3,3);IP.imshow(I2,aspect='auto',cmap="gray");IP.plt.axis("off")
IP.subplot(2,3,4);IP.imshow(I3,aspect='auto',cmap="gray");IP.plt.axis("off")
IP.subplot(2,3,5);IP.imshow(I4,aspect='auto',cmap="gray");IP.plt.axis("off")
IP.subplot(2,3,6);IP.imshow(I5,aspect='auto',cmap="gray");IP.plt.axis("off")
```

## Make a Laplacian Filter

```python
l1 = IP.np.array([[0,1,0],[1,-4,1],[0,1,0]])
I6 = IP.filter2D(isaac,l1,"same")
IP.subplot(1,1,1);IP.imshow(abs(I6),aspect='auto',cmap="gray");IP.plt.axis("off")
```

```python
I88 = IP.np.abs(I6)
```

```python
IP.imshow(isaac+0.9*I88,cmap='gray',aspect="auto")
```

## Sorbel Filtering

```python
W = IP.np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
W
```

```python
I8 = IP.filter2D(isaac,W,"same")
```

```python
IP.imshow(isaac+0.7*abs(I8),cmap="gray",aspect="auto")
```

```python
IP.subplot(1,3,1);IP.imshow(isaac,cmap="gray");IP.plt.axis("off")
IP.subplot(1,3,2);IP.imshow(isaac+0.9*I88,cmap="gray");IP.plt.axis("off")
IP.subplot(1,3,3);IP.imshow(isaac+0.7*abs(I8),cmap="gray");IP.plt.axis("off")
```
